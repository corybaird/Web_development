# Create sqlite3 database

```python
import sqlite3
conn = sqlite3.connect('japan_data.db')
c = conn.cursor()
```

## Import data to df

```python
import pandas as pd
url = "https://www.customs.go.jp/toukei/suii/html/data/d41ma.csv"
df = pd.read_csv(url, encoding = 'unicode_escape', header=2, parse_dates= True, index_col=0).dropna()
df = df[(df.T != 0).any()] #Drops rows of zero
```

## df$\rightarrow$ sql db

```python
df.to_sql('Japan_data',
con = conn
)
```

## Search database

```python
c.execute('''  
SELECT * FROM Japan_data
          ''')
for row in c.fetchall():
    print (row)
```

## sql $\rightarrow$ df

```python
pd.read_sql("select * from Japan_data", conn)
```
